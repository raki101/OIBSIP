```python
import pandas as  pd;
import numpy as np;
import seaborn as sns;
import matplotlib.pyplot as plt;
```

```python
data = pd.read_csv('/content/Unemployment in India.csv');
data.head()
```

```python
data = data.dropna()
data
```

```python
data.describe()
```

```python
data.columns
```

```python
sns.heatmap(data.corr())
```

```python
data = data.rename(columns={data.columns[0]:'State',data.columns[2]:'Frequency',data.columns[3]:'EUR',data.columns[4]:'EE',
                            data.columns[5]:'ELPR',data.columns[6]:'Area'})
data.head()
```

```python
plt.title("Unemployment rate in India");
sns.histplot(x="EE", hue="Area", data = data);
plt.show();
```

```python
sns.violinplot(data=data , x='EE')
```
